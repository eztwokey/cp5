### Постройте столбчатую диаграмму, которая изображает распределение суши Земли по частям света.

Азия – 29,5%

Африка – 20,3%

Северная Америка – 16,2%

Южная Америка – 12%

Антарктида – 9%

Европа – 7%

Австралия – 6%

```python
import matplotlib
import numpy as np
vals = [16, 12, 20, 9, 29, 7, 6]

labels = ["North america", "South America", "Africa", "Antarctica", "Asia", "Europe", "Australia"]
explode = (0.1, 0, 0.15, 0, 0, 0, 0)

fig, ax = plt.subplots()

ax.pie(vals, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode,
wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, rotatelabels=True)
ax.axis("equal")
```

#### С противоположных концов катка длиной 120 м бегут навстречу друг другу
#### два мальчика. Через сколько секунд они встретятся, если второй начнет бег через 5 секунд
#### после первого и если первый пробегает 6 м/с, а второй – 9 м/с? [2].

#### РЕШЕНИЕ
Пусть х с.– время движения второго
мальчика, тогда время первого – (х+5) с. Составим
уравнение, учитывая, что сумма расстояний равна 120
м:6(х + 5) + 9х = 120 решив которое, получим, что х=
6 с. Тогда время, через которое они встретятся – 11 с.

```python
import matplotlib.pyplot as plt
plt.plot([120, 100,80, 60, 40, 20,10])
plt.plot([0, 5, 10, 15, 18, 20, 25])
plt.ylabel('x, м', fontsize=12, color='blue')
plt.xlabel('t, с', fontsize=12, color='blue')
plt.text(4.8, 270, '')
plt.text(4.8, 35, '')
plt.grid()
plt.show()
```
